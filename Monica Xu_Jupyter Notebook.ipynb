```python
# 【4】Univariate Plots
import pandas as pd
import matplotlib.pyplot as plt

# Load Artists.csv into a pandas DataFrame
artists_data = pd.read_csv('/Users/xumeng/Desktop/EDA final project/Artists.csv')

# Display the first few rows of the Artists DataFrame to verify the import
print("Artists Data:")
print(artists_data.head())

# Extract the 'BeginDate' column from the Artists DataFrame
birth_years = artists_data['BeginDate']

# Create a histogram to visualize the distribution of birth years
plt.figure(figsize=(10, 6))
plt.hist(birth_years, bins=30, color='skyblue', edgecolor='black', range=(1750, 2000))  # 设置x轴的范围
plt.title('Distribution of Artist Birth Years')
plt.xlabel('Begin Date')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the histogram
plt.show()
```

```python
# The histogram displayed above represents the distribution of birth years of artists. Here's the analysis based on the provided histogram:

# - X-axis (Begin Date): The x-axis represents the range of artist birth years, spanning from 1750 to 2000.

# - Y-axis (Frequency): The y-axis represents the frequency or count of artists born in each specified year range.

# - Histogram Shape: The histogram is skewed to the right, indicating that there is a higher concentration of artists born in more recent years (towards the right side of the graph). This suggests that there has been a significant increase in the number of artists born in the latter part of the 20th century and early 21st century.

# - Peak Years: The histogram shows one or more peaks, indicating years or periods when a relatively large number of artists were born. In this case, it appears that the peak years are around the mid-20th century, which might be a period of increased artistic activity.

# - Sparse Distribution: Prior to the mid-20th century, the distribution of artist birth years is relatively sparse, with fewer artists born during those earlier periods.

# Overall, this histogram provides insights into the distribution and trends in artist birth years, highlighting the changing patterns of artistic contributions over time. It suggests that the art world has seen a notable increase in artists born in recent decades, possibly reflecting evolving artistic movements and opportunities.
```

```python
#【4】# Bivariate Plots
# Bar plot of top 10 artists' nationalities
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Artists.csv into a pandas DataFrame
artists_data = pd.read_csv('/Users/xumeng/Desktop/EDA final project/Artists.csv')

# Select the top 10 nationalities
top_nationalities = artists_data['Nationality'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 nationalities
filtered_artists_data = artists_data[artists_data['Nationality'].isin(top_nationalities)]

# Create a clustered bar chart
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_artists_data, x='Nationality', hue='Gender', order=top_nationalities)
plt.title('Clustered Bar Chart of Top 10 Artists\' Nationalities by Gender')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# This plot is a clustered bar chart that displays information about the top 10 artists' nationalities and their gender distribution. Here is an analysis of the information presented in the chart:

# 1. X-Axis: The x-axis of the chart represents the "Nationality" of the artists. It displays the top 10 nationalities, and they are labeled on the x-axis.

# 2. Y-Axis: The y-axis represents the count of artists belonging to each nationality. It measures how many artists are from each of the top 10 nationalities.

# 3. Color Differentiation: The bars are further clustered by color to represent the gender of the artists. The legend provides the color-coding for the "Gender" categories.

# 4. Interpretation:
#    - By looking at the chart, you can easily compare the distribution of artists by nationality.
#    - Each nationality has two bars (one for each gender), and the height of each bar represents the count of artists.
#    - It allows you to see how many male and female artists are from each of the top 10 nationalities.
#    - You can also observe any gender imbalances within specific nationalities.

# Overall, this chart provides insights into the gender distribution among artists from the top 10 nationalities, which can be valuable for understanding diversity and representation within the art world.
```

```python
#【4】# Line plot
import pandas as pd
import matplotlib.pyplot as plt

# Load Artists.csv into a pandas DataFrame
artists_data = pd.read_csv('/Users/xumeng/Desktop/EDA final project/Artists.csv')

# Load Artworks.csv into a pandas DataFrame
artworks_data = pd.read_csv('/Users/xumeng/Desktop/EDA final project/Artworks.csv')

# Convert DateAcquired to datetime format, handle parsing errors
artworks_data['DateAcquired'] = pd.to_datetime(artworks_data['DateAcquired'], errors='coerce')

# Group artworks by year of acquisition and count the number of acquisitions per year
acquisition_counts = artworks_data.groupby(artworks_data['DateAcquired'].dt.year)['DateAcquired'].count()

# Create a line plot
plt.figure(figsize=(12, 6))
acquisition_counts.plot(kind='line', marker='o', linestyle='-')
plt.title('Art Acquisitions Over Time')
plt.xlabel('Year of Acquisition')
plt.ylabel('Number of Art Acquisitions')
plt.grid(True)

# Show the line plot
plt.show()
```

```python
# This plot is a line plot that displays information about the number of art acquisitions over time, specifically by year of acquisition. Here is an analysis of the information presented in the chart:

# Interpretation:
#    - The line plot visually represents the changes in the number of art acquisitions made over the years.
#    - It allows you to identify trends, peaks, and troughs in the acquisition activity.
#    - An upward slope indicates an increase in acquisitions over time, while a downward slope indicates a decrease.
#    - Sudden spikes or drops in the line may indicate significant events or changes in art acquisition patterns.

# Overall, this chart provides insights into the historical pattern of art acquisitions, helping to understand the evolution of the collection over time and identify trends in the acquisition of artworks.
```

```python
# The box plot presented above provides insights into the distribution of artwork dimensions, specifically focusing on the width (in centimeters). Here's an analysis based on the provided box plot:

# - X-axis (Width in cm): The x-axis represents the width of artworks in centimeters. Each box in the plot corresponds to a range of widths, and outliers are depicted as individual data points beyond the "whiskers" of the boxes.

# - Box: The box itself represents the interquartile range (IQR) of artwork widths. It spans from the 25th percentile (Q1) to the 75th percentile (Q3). The box's length signifies the spread of the middle 50% of the data.

# - Whiskers: The whiskers extend from the edges of the box and typically encompass data points within a certain range. They help identify the data's spread beyond the IQR.

# - Outliers: Individual data points lying beyond the whiskers are considered outliers. They are depicted as dots on the plot and may indicate unusual or extreme values.

# Analysis:

# - The box plot allows us to understand the distribution of artwork widths within the specified range of approximately 4000 cm to 5000 cm.

# - The box itself shows the middle 50% of artworks, with the length of the box indicating how spread out these artworks are in terms of width.

# - The whiskers reveal the range within which the majority of artworks fall.

# - Outliers are identified as individual data points that deviate significantly from the typical range of widths.

# In summary, this box plot helps art analysts and enthusiasts understand the distribution and potential outliers in the dimensions (width) of artworks within the specified range, providing valuable insights into the dataset's characteristics.
```
